```python
import os
from utils import data_loader_utils
```

```python
machines = ["M01","M02","M03"]
process_names = ["OP00","OP01","OP02","OP03","OP04","OP05","OP06","OP07","OP08","OP09","OP10","OP11","OP12","OP13","OP14"]
labels = ["good","bad"]
path_to_dataset = "../data/"
```

```python
X_data = []
y_data = []
def process_data():
    for process_name in process_names:
        for machine in machines:
            for label in labels:
                data_path = os.path.join(path_to_dataset, machine, process_name, label)
                data_list, data_label = data_loader_utils.load_tool_research_data(data_path, label=label)
                X_data.append(data_list)
                y_data.append(data_label)
    print(f"X_data:{X_data}")
process_data()
```

```python
import numpy as np

# Get the number of features from the first array in X_data
num_features = X_data[0][0].shape[1]

# Initialize empty lists for each feature
features = [[] for _ in range(num_features)]

# Extract features from each sublist
for arr in X_data:
    for item in arr:
        for i in range(num_features):
            features[i].append(item[i])

# Convert each feature list into a NumPy array
data_arrays = [np.array(feature) for feature in features]
```

```python
data_arrays 
```

```python
def adf_test(data):
    flattened_data = data.flatten()

    result = sm.tsa.stattools.adfuller(flattened_data)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
```

```python
def acf_test(data):
    acf = sm.tsa.stattools.acf(data)
    print('Autocorrelation Function (ACF):', acf)
```

```python
for i, data_array in enumerate(data_arrays):
    print(f"Feature {i+1}:")
    adf_test(data_array)
```
